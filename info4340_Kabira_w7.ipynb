```python
# import libraries
import pandas as pd
from sklearn.datasets import load_iris
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
from factor_analyzer.factor_analyzer import calculate_kmo
```

```python
# read in the personality data
df= pd.read_csv("food-texture.csv")
df.head()
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.info()
```

```python
# Kaiser-Meyer-Olkin (KMO) Test

kmo_all,kmo_model=calculate_kmo(df)
```

```python
kmo_model
```

### The overall KMO for the model is .71, which indicates that the data is suitable for factor analysis.

â€¢	Conduct principal axis component factor analysis with Varimax rotation. Generate your eigenvalues, create a scree plot, and get your factor loadings. Explain your findings for the eigenvalues, scree plot, and factor loadings from the model. What would you change?

```python
# create FA object with principal method and varimax rotation
fa_principal = FactorAnalyzer(rotation = 'varimax', method = 'principal')
fa_principal.fit(df)

loadings = fa_principal.loadings_
loadings
```

```python
# Make array look better
loadings = pd.DataFrame(fa_principal.loadings_, columns=['Factor 1', 'Factor 2', 'Factor 3'], index = df.columns)
print('Factor Loadings \n%s' %loadings)
```

```python
# check Eigenvalues
eigen_values_principal, vectors = fa_principal.get_eigenvalues()
eigen_values_principal
```

```python
# create scree plot
plt.scatter(range(1, df.shape[1]+1), eigen_values_principal)
plt.plot(range(1, df.shape[1]+1), eigen_values_principal)
plt.title('Principal Method Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

### I find this analysis unsatisfying. The Factor Loadings seem rather off.

### The Eigenvalues also indicate that Factors 1 and 2 are significant sice both have Eigenvalues greater than 1, and Factor 3's Eigenvalue is less than 1/2.

### The Scree Plot shows the Eigenvalues visually; even though Factor 3 is the elbow (and therefore worth consideration), it is so much lower than Factor 2, it seems worth excluding.

### BUT, according to the Factor Loadings, Factor 3 explains most of the variance in Oil and Density. In fact, the value for Oil is nearly as large as the value of Hardness for Factor 2.  The amount of variance that Factor 3 explains leads me to expect that Factor 3's Eigenvalue would be much higer than .3.
```python
# create FA object with minres method and varimax rotation
fa_minres = FactorAnalyzer(rotation = 'varimax')  #Minres is default and therefore not specified
fa_minres.fit(df)

loadings1 = fa_minres.loadings_
loadings1
```

```python
# Make array look better
loadings1 = pd.DataFrame(fa_minres.loadings_, columns=['Factor 1', 'Factor 2', 'Factor 3'], index = df.columns)
print('Factor Loadings \n%s' %loadings1)
```

```python
# check Eigenvalues
eigen_values_minres, vectors = fa_minres.get_eigenvalues()
eigen_values_minres
```

```python
# create scree plot
plt.scatter(range(1, df.shape[1]+1), eigen_values_minres)
plt.plot(range(1, df.shape[1]+1), eigen_values_minres)
plt.title('Minres Method Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

### This analysis makes more sense, with the Factor Loadings values supporting the Eigenvalues. The Eigenvalues for both this analysis and the previous one are identical.

### The Factor Loadings suggest that two factors are the most significant.  From the loading array, the data suggests that Factor 1 explains the vast majority of the variance in Oil and Density; and the majority of the variance for Crispy and Fracture. Factor 2 explains most of the remaining variance for Crispy and Fracture and the majority of the variance of Hardness.  Factor 3, with its values under .15 explains almost none of the variance for any of the factors.

### The Eigenvalues also indicate that Factors 1 and 2 are significant sice both have Eigenvalues greater than 1, and Factor 3's Eigenvalue is less than 1/2.

### Because the Minres method delivered Factor Loadings that look consistent with the Eigenvalues, I choose to go forward with that model.

```python
# repeating factor loadings for FA object with minres method, varimax rotation, and only 2 factors
fa_minres_2 = FactorAnalyzer(2, rotation = 'varimax')
fa_minres_2.fit(df)

loadings2 = fa_minres_2.loadings_
loadings2
```

```python
# Make factor loadings array look better
loadings2 = pd.DataFrame(fa_minres_2.loadings_, columns=['Factor 1', 'Factor 2'], index = df.columns)
print('Factor Loadings \n%s' %loadings2)
```

### From the 2-factor loading array, the data suggests that Factor 1 continues to explain the vast majority of the variance in Oil and Density; and the majority of the variance for Crispy and Fracture. Factor 2 continues to explain most of the remaining variance for Crispy and Fracture and the majority of the variance of Hardness.

```python
# obtaining variance of each factors
fa_minres_2.get_factor_variance()
```

```python
# Make variance array look better
fa_minres_2_var = pd.DataFrame(fa_minres_2.get_factor_variance(), columns=['Factor 1', 'Factor 2'], 
                           index= ['Loadings', 'Proportional Var', 'Cumulative Var'])
print('Factor Variance \n%s' %fa_minres_2_var)
```

### The Loadings and Proportional Variance suggest that Factor 1 describes twice the amount of variance as does Factor 2. Together they explain more than 76% of the total variance.

```python
# obtaining communalities for factors
fa_minres_2.get_communalities()
```

```python
# showing communalities
fa_minres_2_comm = pd.DataFrame(fa_minres_2.get_communalities(), index=df.columns, columns=['Communalities'])
print('Factor Communalities \n%s' %fa_minres_2_comm)
```

### The Factor Communalities suggest that Crispy has the highest amount of its variance explained by the model at 95%, followed by Density at 83% and Fracture at 74%.   

```python
# obtaining uniqueness for factors
fa_minres_2.get_uniquenesses()
```

```python
# showing uniqueness
fa_minres_2_uni = pd.DataFrame(fa_minres_2.get_uniquenesses(), index=df.columns, columns=['Uniquenesses'])
print('Factor Uniquenesses \n%s' %fa_minres_2_uni)
```

### Since Uniqueness is measured as 1-Communality, the model indicates that almost 40% of the variance of Hardness and 32% of the variance of Oil remain unexplained by the model. Another way to think about this is that Hardness and Oil are the variables that are least important to the model as a whole.

#### In analytics, we like to describe data in the simplest way that still captures its essence.  One method we use for this is called exploratory factor analysis. 

#### I think of it like distilling a photograph into a sketch that preserves the most important elements.  The elements are our factors. Let's say you're at a museum that doesn't allow photography, and you want to show someone a picture that moved you.

#### Look at the photograph below. How much detail do you need to convey its essence?

#### We try various ways of simplifying data. This is the exploratory part.  Sometimes we try something too simple...  
#### While we see this is likely a woman and that she has her arms crossed and is likely standing in snow, we don't see her hair across her face, which seems an important detail.

#### So we add a bit more detail to our data (in the form of additional factors) to capture enough to tell us enough of what we need to know.
#### We can now guess that she is next to a body of water and that her hair is across her face, suggesting wind. That and the snow and her crossed arms indicate that she is cold. We also have a sense of isolation. 
#### In our exploration of the important factors of this image we have settled on a level of simplicity that still conveys the important meaning.
