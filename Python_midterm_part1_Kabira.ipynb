## Data Exploration
One way I explored the data was to look at spending on discretionary items and to compare that spending to spending on education.  I took a particular look at spending on toys and toy categories because children are generally the biggest recipients of toys and the ones receiving education.

In order to compare spending values across countries I converted all of the local currency values to USD using a Purchasing Power Parity index, which I obtained with the other data. Then I divided all of the indexed purchasing totals by the countries' populations in order to be able to compare values of countries that have smaller populations with those of countries with larger populations.

The scatterplot of education spending vs per capita disposable income gives a graphic view of the positive correlation of these two variables.  The trendline’s slope is moderate, and the correlation matrix provides the value of .41 for this correlation.  A couple of interesting observations from the data are that disposable income is most highly correlated with the amount of money spent on toys (.86) versus all of the other discretionary spending categories.  The discretionary spending category with the lowest correlation to education spending was Reading Materials.  Literacy is a focus of education so the low correlation caught my attention. If we look at toy categories specifically, Baby Toys and Games & Puzzles are the two categories that are negatively correlated (though slightly so) with Education spending (-.2 and -.1, respectively).  The category of toys that is most highly correlated with education spending is video game software.  Video game software is also the only toy category which has a minimum value greater than zero for the dataset.

From a regional standpoint, Australasia and North America rank the highest in education spending.  However, these two regions each represent only two countries: Australia and New Zealand for Australasia and Canada and the US for North America.  Perhaps of more interest are the Middle East and Africa region and the Asia Pacific region.  Not only are these the regions with the most countries, but they also have a wider variety of education spending and disposable income values, including significant outliers. Western Europe slightly lags behind Latin America on education spending, even though its disposable income amounts are twice as large as those of Latin America.

```python
# Reading in Purchasing Power Parity file
df2=pd.read_csv('Midterm_PPP.csv')
df2.head()
```

```python
# Reducing number of columns in PPP dataframe
df2Rev=df2.drop(columns=['Category', 'Data Type', 'Unit', 'Current Constant'])
df2Rev
```

```python
# Checking data types and columns
df3.info()
```

```python
# Reading in toy-specific file
df4=pd.read_csv('Midterm_Toys.csv')
df4.head()
```

```python
# Checking data types and columns
df4.info()
```

```python
# Pivoting columns to make one row per country
df4Pivot=df4.pivot('Country_ID', 'Category', 'Value')
df4Pivot.head()
```

```python
# Merging Region file with big economic category file
dfMerge1=pd.merge(df3, df1Pivot, on='Country_ID', how='left')
```

```python
# Merging in PPP file
dfMerge2=pd.merge(dfMerge1, df2Rev, on='Country_ID', how='left')
dfMerge2.head()
```

```python
# Checking data types and columns
dfMerge2.info()
```

```python
# Merging in toy-specific file
dfMerge3=pd.merge(dfMerge2, df4Pivot, left_on = 'Country_ID', right_index=True, how='left')
dfMerge3.head()
```

```python
# Cleaning up new file with renaming columns and dropping extra country column
dfMerge4=dfMerge3.drop(columns='Geography_y')
dfMerge5=dfMerge4.rename(columns={'Geography_x' : 'Country',
                                  'Consumer Expenditure on Audio-Visual & Photographic and Information Processing Equipment' : 'A/V & Photo',
                                  'Consumer Expenditure on Other Major Durables For Recreation and Culture' : 'Recreation - Other',
                                  'Consumer Expenditure on Other Recreational Items and Equipment & Gardens and Pets' : 'Garden & Pets',
                                  'Consumer Expenditure on Recreational and Cultural Services' : 'Recreational & Cultural Services',
                                  'Consumer Expenditure on Newspapers & Magazines & Books and Stationery' : 'Reading Materials',
                                  'Consumer Expenditure on Education' : 'Education',
                                  'Consumer Expenditure on Personal Care' : 'Personal Care',
                                  'Consumer Expenditure on Jewellery & Silverware & Watches and Clocks & Travel Goods' : 'Jewellery & Travel Goods',
                                  'Index' : 'PPP Index',
                                  'Action Figures and Accessories' : 'Action Figure Toys',
                                  'Arts and Crafts' : 'Arts & Crafts Toys',
                                  'Baby and Infant' : 'Baby Toys',
                                  'Construction' : 'Construction Toys',
                                  'Dolls and Accessories' : 'Dolls',
                                  'Dress-Up and Role Play' : 'Dress-Up & Role Play',
                                  'Games and Puzzles' : 'Games & Puzzles',
                                  'Other Traditional Toys and Games' : 'Other Toys & Games',
                                  'Outdoor and Sports' : 'Outdoor Toys & Sports',
                                  'Plush' : 'Plush Toys',
                                  'Pre-School' : 'Pre-School Toys',
                                  'Ride-On Vehicles' : 'Ride-On Toy Vehicles',
                                  'Scientific/Educational' : 'Educational Toys'})
dfMerge5.head()
                                  
```

```python
# Checking data types and columns
dfMerge5.info()

# No null values for major economic measures; only 32 countries had toy data so all others are null for these variables
```

```python
# Recoding the Region variable with friendly names
dfMerge5['Region']=dfMerge5['Region_Code'].replace(to_replace=['AP','AA','EE', 'LA', 'MEA', 'NOA', 'WE'],
                                             value=['Asia Pacific','Australasia','Eastern Europe',
                                             'Latin America', 'Middle East and Africa', 
                                             'North America', 'Western Europe'])
```

```python
# Standardizing money amounts by dividing local currencies by PPP index and changing values (were in millions) to per capita by 
# dividing by population (is in thousands).  

# The resulting purchase values are as written; Disposable income is in thousands

dfMerge5['A/V & Photo Adj'] = dfMerge5['A/V & Photo'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Education Adj'] = dfMerge5['Education'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Jewellery & Travel Goods Adj'] = dfMerge5['Jewellery & Travel Goods'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Reading Materials Adj'] = dfMerge5['Reading Materials'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Recreation - Other Adj'] =dfMerge5['Recreation - Other'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Garden & Pets Adj'] = dfMerge5['Garden & Pets'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Personal Care Adj'] = dfMerge5['Personal Care'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Recreational & Cultural Services Adj'] = dfMerge5['Recreational & Cultural Services'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Disposable Income Adj'] = dfMerge5['Disposable Income'] / dfMerge5['PPP Index'] / dfMerge5['Total Population']
dfMerge5['Action Figure Toys Adj'] = dfMerge5['Action Figure Toys'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Arts & Crafts Toys Adj'] = dfMerge5['Arts & Crafts Toys'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Baby Toys Adj'] = dfMerge5['Baby Toys'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Construction Toys Adj'] = dfMerge5['Construction Toys'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Dolls Adj'] = dfMerge5['Dolls'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Dress-Up & Role Play Adj'] = dfMerge5['Dress-Up & Role Play'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Games & Puzzles Adj'] = dfMerge5['Games & Puzzles'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Model Vehicles Adj'] = dfMerge5['Model Vehicles'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Other Toys & Games Adj'] = dfMerge5['Other Toys & Games'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Outdoor Toys & Sports Adj'] = dfMerge5['Outdoor Toys & Sports'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Plush Toys Adj'] = dfMerge5['Plush Toys'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Pre-School Toys Adj'] = dfMerge5['Pre-School Toys'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Remote Control Toys Adj'] = dfMerge5['Remote Control Toys'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Ride-On Toy Vehicles Adj'] = dfMerge5['Ride-On Toy Vehicles'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Educational Toys Adj'] = dfMerge5['Educational Toys'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Video Games Hardware Adj'] = dfMerge5['Video Games Hardware'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Video Games Software Adj'] = dfMerge5['Video Games Software'] / dfMerge5['PPP Index'] / dfMerge5['Total Population'] * 1000
dfMerge5['Total Toys'] = dfMerge5['Action Figure Toys Adj']+dfMerge5['Arts & Crafts Toys Adj']+dfMerge5['Baby Toys Adj']+dfMerge5['Construction Toys Adj']+dfMerge5['Dolls Adj']+dfMerge5['Dress-Up & Role Play Adj']+dfMerge5['Games & Puzzles Adj'] +dfMerge5['Model Vehicles Adj']+dfMerge5['Other Toys & Games Adj']+dfMerge5['Outdoor Toys & Sports Adj']+dfMerge5['Plush Toys Adj']+dfMerge5['Pre-School Toys Adj']+dfMerge5['Remote Control Toys Adj']+dfMerge5['Ride-On Toy Vehicles Adj']+dfMerge5['Educational Toys Adj']+dfMerge5['Video Games Hardware Adj']+dfMerge5['Video Games Software Adj']
```
